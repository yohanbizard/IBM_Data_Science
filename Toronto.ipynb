```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python
data_source = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(data_source).text
soup = BeautifulSoup(source, 'xml')
table=soup.find('table')
```

```python
#Create the dataframe columns
column_names = ['Postcode','Borough','Neighborhood']
df = pd.DataFrame(columns = column_names)
```

```python
df
```

```python
for tr_cell in table.find_all('tr'):
    row_data=[]
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
    if len(row_data)==3:
        df.loc[len(df)] = row_data
```

```python
df
```

```python
#Clean dataframe from Not assigned values
df=df[df['Borough']!='Not assigned']
```

```python
# Merging borough with the same postcode
df2 = df.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Assign borought name when neighborhood is Not assign
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])
```

```python
df2
```

```python
df2.shape
```

```python
#Importing latitute and longitude data
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
lat_lon.rename(columns={'Postal Code':'Postcode'},inplace=True)
neigh_data = pd.merge(df2,lat_lon,on='Postcode')
neigh_data.head()
```

```python
#Filter to keep only Toronto data
neigh_data_toronto = neigh_data[neigh_data['Borough'].str.contains('Toronto',regex=False)]
```

```python
neigh_data_toronto
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(neigh_data_toronto['Borough'].unique()),
        neigh_data_toronto.shape[0]
    )
)
```

```python
# create map of New York using latitude and longitude values

map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)
# add markers to map
for lat,lng,borough,neighborhood in zip(neigh_data_toronto['Latitude'],neigh_data_toronto['Longitude'],neigh_data_toronto['Borough'],neigh_data_toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
#Using KMeans clustering for the clustering of the neighbourhoods
```

```python
k=5
toronto_clustering = neigh_data_toronto.drop(['Postcode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
neigh_data_toronto.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
neigh_data_toronto
```

```python
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighborhood, cluster in zip(neigh_data_toronto['Latitude'], neigh_data_toronto['Longitude'], neigh_data_toronto['Neighborhood'], neigh_data_toronto['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
